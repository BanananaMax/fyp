```python
c = tweepy.Cursor(api.search, q="bitcoin", include_entities=False, until=yesterday).items()

counter =1
while True:
    try:
        tweet = c.next()
        twt = {
            "created": tweet._json.created_at,
            "id": tweet._json.id,
            "user": tweet._json.user.id,
            "text": tweet._json.text
        }
        if( counter ==1):
            print(twt)

            
        counter = counter + 1
        # Insert into db
    except tweepy.TweepError:
        time.sleep(60 * 15)
        continue
    except StopIteration:
        break
```
